```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

threads = ["1", "2", "4", "8", "16", "32", "64", "128"]
tempos = [65.387, 37.376, 24.062, 27.034, 26.046, 28.028, 27.376, 26.693]

# Adicionando os valores exatos acima de cada barra
for i, tempo in enumerate(tempos):
    ax.text(i, tempo + 1, f'{tempo:.2f}', ha='center', va='bottom')

ax.bar(threads, tempos, color='red')

ax.set_ylabel('Tempo (s)')
ax.set_title('Comparação entre quantidade de threads no loop externo (i)')

plt.show()
```
