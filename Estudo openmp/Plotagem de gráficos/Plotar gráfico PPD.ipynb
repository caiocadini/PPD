```python
import matplotlib.pyplot as plt
import numpy as np

threads = ("1", "2", "4", "8", "16", "32", "64", "128")
tempos = {
    'For externo (i)': (65.387, 37.376, 24.062, 27.034, 26.046, 28.028, 27.376, 26.693),
    'For intermedirio (j)': (63.706, 34.029, 34.035, 32.422, 33.369, 34.048, 33.082, 34.061),
    'For interno (k)': (68.044, 32.364, 32.737, 119.351, 458.042, 833.381, 0, 0),
}

x = np.arange(len(threads))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in tempos.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Tempo (s)')
ax.set_title('Comparao entre paralelizaes em diferentes loops')
ax.set_xticks(x + width, threads)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 950)

plt.show()
```
